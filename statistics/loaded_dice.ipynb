```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random samples
numbers = [1, 2, 3, 4, 5, 6]
size=10000000
results = np.random.choice(numbers,size=size)

# Manually count occurrences
counts = {number: 0 for number in numbers}
print(counts)
for n in results:
    counts[n] += 1

# Prepare data for plotting
x_values = list(counts.keys())
y_values = list(counts.values())
y_values = [value / size  for value in y_values]
print(y_values)

# Plotting the histogram
plt.bar(x_values, y_values, edgecolor='black', color='skyblue')
plt.xlabel('Number')
plt.ylabel('Frequency')
plt.title('Histogram of Randomly Chosen Numbers')
plt.show()
print(counts)

print("the expected value is")
expected_value =np.sum(np.array(x_values)*np.array(y_values))
print(expected_value)
variance =  np.sum(np.array(y_values) * (np.array(x_values) -expected_value)**2)
print("the variance is")
print(variance)
```
